```python
with open("alice_in_wonderland.txt", "r") as file:
    book = file.read()
```

# Find Number of chapters

```python
import re

pattern = re.compile("CHAPTER [A-Z]+")
findings = re.findall(pattern, book)
findings
```

# How many times "Alice" is called?

```python
pattern = re.compile("Alice")
findings = re.findall(pattern, book)
len(findings)
```

# Wich are the sentences where "carefully" was used?

```python
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+carefully[^a-zA-Z]+[^.]*")
findings = re.findall(pattern, book.lower())
findings = [finding.replace("\n", "") for finding in findings]
findings
```

# What is the most used words? 

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
findings
```

```python
d = {}

for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    
    else:
        d[word] = 1
```

```python
d.keys()
```

```python
e = d.items()
d_list = [(value, key) for (key, value) in d.items()]
sorted(d_list, reverse=True)
```

```python
import nltk

from nltk.corpus import stopwords
nltk.download('stopwords')
english_stopwords = stopwords.words("english")
```

```python
filtered_words = []
for count, word in d_list:
    if word not in english_stopwords:
        filtered_words.append((count, word))
sorted(filtered_words, reverse=True)
```
